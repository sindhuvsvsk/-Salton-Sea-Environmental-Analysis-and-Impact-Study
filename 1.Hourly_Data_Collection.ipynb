```python
import pandas as pd
from glob import glob
import os
```

```python
Pollutants=['CO','NO2','OZONE','PM10','PM25HR','SO2']
base_location=r'C:\Users\samik\Documents\Fall2021\New_Data_Approach\New_Data'
```

```python
filename_dict={} #dictionary to store filenames with key as pollutant
for i in range(len(Pollutants)):
    files=glob(os.path.join(base_location,Pollutants[i]+'_*.csv'))
    filename_dict[Pollutants[i]]=files

filename_dict['CO']
```

```python
#Code to merge the individual year files of pollutants into one, creating total of 6 dataframes, one for each pollutant
for i,j in filename_dict.items():
    print(i)
    if i=='CO':
        dataframes_CO = [pd.read_csv(f,skipfooter=11, engine='python') for f in j] 
        CO_combined = pd.concat(dataframes_CO, axis=0, ignore_index=True)
    elif i=='NO2':
        dataframes_NO2 = [pd.read_csv(f,skipfooter=11, engine='python') for f in j] 
        NO2_combined = pd.concat(dataframes_NO2, axis=0, ignore_index=True)
    elif i=='OZONE':
        dataframes_OZONE = [pd.read_csv(f,skipfooter=11, engine='python') for f in j] 
        OZONE_combined = pd.concat(dataframes_OZONE, axis=0, ignore_index=True)
    elif i=='PM10':
        dataframes_PM10 = [pd.read_csv(f,skipfooter=11, engine='python') for f in j] 
        PM10_combined = pd.concat(dataframes_PM10, axis=0, ignore_index=True)
    elif i=='PM25HR':
        dataframes_PM25HR = [pd.read_csv(f,skipfooter=11, engine='python') for f in j] 
        PM25HR_combined = pd.concat(dataframes_PM25HR, axis=0, ignore_index=True)
    elif i=='SO2':
        dataframes_SO2 = [pd.read_csv(f,skipfooter=11, engine='python') for f in j] 
        SO2_combined = pd.concat(dataframes_SO2, axis=0, ignore_index=True)
```

```python
#count of all the combined raw datasets
print("CO count: ",CO_combined.shape)
print("NO2 count: ",NO2_combined.shape)
print("Ozone Count: ",OZONE_combined.shape)
print("PM10 count: ",PM10_combined.shape)
print("PM25 count: ",PM25HR_combined.shape)
print("SO2 count: ",SO2_combined.shape)
```

```python
# Perform same operation for meteorological data.
Meteor=['BPSTA','DEWPNT','SWINSPD','WINSPD','RELHUM','TEMP']
base_location_m=r'C:\Users\samik\Documents\Fall2021\New_Data_Approach\New_Data\Weather'
```

```python
wind_speed=pd.read_csv('../New_Data/Weather/Wind_data.csv')
```

```python
filename_mt_dict={} #dictionary to store filenames with key as pollutant
for i in range(len(Meteor)):
    files=glob(os.path.join(base_location_m,Meteor[i]+'_*.csv'))
    filename_mt_dict[Meteor[i]]=files

filename_mt_dict['TEMP']
```

```python
#Code to merge the individual year files of pollutants into one, creating total of 6 dataframes, one for each pollutant
for i,j in filename_mt_dict.items():
    print(i)
    if i=='BPSTA':
        dataframes_BPSTA = [pd.read_csv(f,skipfooter=11, engine='python') for f in j] 
        BPSTA_combined = pd.concat(dataframes_BPSTA, axis=0, ignore_index=True)
    elif i=='DEWPNT':
        dataframes_DEWPNT = [pd.read_csv(f,skipfooter=11, engine='python') for f in j] 
        DEWPNT_combined = pd.concat(dataframes_DEWPNT, axis=0, ignore_index=True)
    elif i=='RELHUM':
        dataframes_RELHUM = [pd.read_csv(f,skipfooter=11, engine='python') for f in j] 
        RELHUM_combined = pd.concat(dataframes_RELHUM, axis=0, ignore_index=True)
    elif i=='TEMP':
        dataframes_TEMP = [pd.read_csv(f,skipfooter=11, engine='python') for f in j] 
        TEMP_combined = pd.concat(dataframes_TEMP, axis=0, ignore_index=True)
```

```python
#count of all the combined raw datasets
print("Barometric Pressure count: ",BPSTA_combined.shape)
print("DewPoint count: ",DEWPNT_combined.shape)
print("Wind Count: ",wind_speed.shape)
print("Relative Humidity count: ",RELHUM_combined.shape)
print("Temperature count: ",TEMP_combined.shape)
```

### Group the data in each file on date and hour. So we will have one row per pollutant for an hour. Dataset consists data from one site but grouping will ensure we have data at single site level around Salton Sea.

```python
# Grouping for each date and hour
grouped_CO = CO_combined.groupby(['date', 'start_hour'])
grouped_NO2= NO2_combined.groupby(['date', 'start_hour'])
grouped_OZONE=OZONE_combined.groupby(['date', 'start_hour'])
grouped_PM10 = PM10_combined.groupby(['date', 'start_hour'])
grouped_PM25HR= PM25HR_combined.groupby(['date', 'start_hour'])
grouped_SO2=SO2_combined.groupby(['date', 'start_hour'])
```

```python
CO_combined
```

```python
# create a new data frame 
d2_CO = grouped_CO['value'].max().to_frame(name = 'CO').reset_index()
d2_NO2= grouped_NO2['value'].max().to_frame(name = 'NO2').reset_index()
d2_OZONE= grouped_OZONE['value'].max().to_frame(name = 'OZONE').reset_index()
d2_PM10= grouped_PM10['value'].max().to_frame(name = 'PM10').reset_index()
d2_PM25HR= grouped_PM25HR['value'].max().to_frame(name = 'PM25').reset_index()
d2_SO2= grouped_SO2['value'].max().to_frame(name = 'SO2').reset_index()
```

```python
#count of all the combined raw datasets
print("CO count: ",d2_CO.shape)
print("NO2 count: ",d2_NO2.shape)
print("Ozone Count: ",d2_OZONE.shape)
print("PM10 count: ",d2_PM10.shape)
print("PM25 count: ",d2_PM25HR.shape)
print("SO2 count: ",d2_SO2.shape)
```

```python
#meteorological data
# Grouping for each date and hour
grouped_BPSTA = BPSTA_combined.groupby(['date', 'start_hour'])
grouped_DEWPNT= DEWPNT_combined.groupby(['date', 'start_hour'])
grouped_Wind=wind_speed.groupby(['date', 'start_hour'])
grouped_RELHUM = RELHUM_combined.groupby(['date', 'start_hour'])
grouped_TEMP= TEMP_combined.groupby(['date', 'start_hour'])
```

```python
# create a new grouped data frame 
d2_BP = grouped_BPSTA['value'].mean().to_frame(name = 'Barometric_Pressure').reset_index()
d2_DP= grouped_DEWPNT['value'].mean().to_frame(name = 'Dewpoint').reset_index()
#d2_WIND= grouped_Wind['value'].mean().to_frame(name = 'Wind').reset_index()
d2_RH= grouped_RELHUM['value'].mean().to_frame(name = 'Relative_Humidity').reset_index()
d2_TEMP= grouped_TEMP['value'].mean().to_frame(name = 'Temperature').reset_index()
```

```python
#count of all the combined raw datasets
print("Barometric Pressure count: ",d2_BP.shape)
print("DewPoint count: ",d2_DP.shape)
print("Relative Humidity count: ",d2_RH.shape)
print("Temperature count: ",d2_TEMP.shape)
```

### Merge all the grouped pollutant files into one based on date and hour.

```python
#data["Date"]= pd.to_datetime(data["Date"]) code for changing into date object.

#Merge all the six dataframes using date and hour. Perform outer join.
Air_pollutants_hourly=d2_CO.merge(d2_NO2,on=['date','start_hour'],how='outer').merge(d2_OZONE,on=['date','start_hour'],how='outer').merge(d2_PM10,on=['date','start_hour'],how='outer').merge(d2_PM25HR,on=['date','start_hour'],how='outer').merge(d2_SO2,on=['date','start_hour'],how='outer')
```

```python
Air_pollutants_hourly.index = (pd.to_datetime(Air_pollutants_hourly['date']) + pd.to_timedelta(Air_pollutants_hourly['start_hour'], unit='h'))
```

```python
Air_pollutants_hourly.sort_index(inplace=True)
```

```python
#Merge all the 5 dataframes using date and hour. Perform outer join.
Meteor_hourly=d2_BP.merge(d2_DP,on=['date','start_hour'],how='outer').merge(d2_RH,on=['date','start_hour'],how='outer').merge(d2_TEMP,on=['date','start_hour'],how='outer').merge(wind_speed,on=['date','start_hour'],how='outer')
```

```python
#Finally create one single dataset with air pollutants and meteorological data frames and using to_csv to store into a csv file
Hourly_Dataset=Air_pollutants_hourly.merge(Meteor_hourly,on=['date','start_hour'],how='outer')
```

```python
Hourly_Dataset.index = (pd.to_datetime(Hourly_Dataset['date']) + pd.to_timedelta(Hourly_Dataset['start_hour'], unit='h'))
Hourly_Dataset.sort_index(inplace=True)
```
